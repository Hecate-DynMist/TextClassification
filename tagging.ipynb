```python
import pandas as pd
eign = pd.read_excel('Eigen56.xlsx')
article = pd.read_excel('Articles56.xlsx')
daily = pd.read_excel('Dailies56.xlsx')
label = pd.read_excel('label.xlsx')
```

```python
article = article[['id','title','description','content']]

import re
def clean(raw):
    r1 = u'[a-zA-Z0-9’!"#$%&\'()*+,-./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~\s（）：△ ©]+'
    r2 = "[\.\!\/_,$%^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]+"
    r3 =  "[.!//_,$&%^*()<>+\"'?@#-|:~{}]+|[——！\\\\，。=？、：“”‘’《》【】￥……（）△:]+" 
    r4 =  "\\【.*?】+|\\《.*?》+|\\#.*?#+|[.!/_,$&%^*()<>+""'?@|:~{}#]+|[——！\\\，。=？、：“”‘’￥……（）《》【】]"

    text1 = re.sub(r1,'',raw)
    text2 = re.sub(r2,'',text1)
    text3 = re.sub(r3,'',text2)
    text4 = re.sub(r4,'',text3)
    return text4
```

```python
def septext(text):
    Lword = []
    for x in text: 
        Lword.append(''.join(x))
    with open('/home/lene/Documents/chinese_stop_words.txt') as f:
        stopWords = [line.strip() for line in f.readlines()]
    stopWords.extend(['跟', '后', '在','再次','多个','的','另','被','年','月','而','与','所以','这些', \
                      '近年来','我','文曾','至','以','去年','为','已','人'])
    words = [w for w in Lword if w not in stopWords]
    return words
```

```python
import jieba
import warnings
warnings.filterwarnings('ignore')

def token(df):
    df['ccontent'] = ''
    df['ctitle'] = ''
    df.content = df.content.astype(str)
    df.title = df.title.astype(str)

    for i in range(len(df)):
        df.ccontent[i] = septext(jieba.cut(clean(df.content[i])))
        df.ctitle[i] = septext(jieba.cut(clean(df.title[i])))
```

```python
from ast import literal_eval
article = article.drop_duplicates(subset=['title'])
tag = pd.read_excel('article_tag_hecate.xlsx')
tag =tag[['title','name']]
df = pd.merge(article,tag,on='title',how='left')

def transfername(df):
    for i in range(len(df)):
        if str(df['name'][i])!='nan':
            df['name'][i] = set(literal_eval(df['name'][i]))
    return df
```

```python
def match(words):
    tag=set()
    for column in label.columns:
        for word in words:
            if word in label[column].tolist():
                tag.add(str(column))
    return tag
```

```python


df['tag_title'] = ''
df['tag_name'] = ''
df['tag_content'] = ''
def tag(df,base,tag):
    for i in range(len(df)):
        if str(df[base][i])!='nan':
            df[tag][i] = match(df[base][i])
    return df
tag(df,'ctitle','tag_title')
tag(df,'name','tag_name')
tag(df,'ccontent','tag_content')
```

```python
df.to_excel('article_tagged.xlsx')
```
